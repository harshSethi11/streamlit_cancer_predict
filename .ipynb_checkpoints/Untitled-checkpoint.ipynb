# Linear Regression 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
pd.set_option('display.width', 1000)
data = pd.read_csv("housing.csv",sep = r'\s+', header = None)
```

```python
print(f"Data Shape: {data.shape}")
print(data.head())
```

```python
data = data.apply(pd.to_numeric, errors='coerce')  
print(data.head())
```

```python
data = data.fillna(data.mean())
print(data.head())
```

```python
print("Data after filling Nans with column mean")
print(data.describe())
```

```python
if data.empty:
    raise ValueError("the dataset is empty after handling NAN values")
```

```python
X= data.iloc[:,:-1].values
y= data.iloc[:,-1].values.reshape(-1,1)
scaler = StandardScaler()
X = scaler.fit_transform(X)  # Normalize feature matrix
#y = scaler.fit_transform(y)
print(X)
print(y)
```

```python
bias = 0
m=len(y)
print(m)
lr = 0.0004
iterations =1000
```

```python
weights = np.zeros((X.shape[1],1))
print(f"shape of weights: {weights.shape}")
```

```python
print(y.shape)
```

```python
def compute_cost(X,y,weights,bias):
    predictions = X.dot(weights) + bias
    error = predictions -y
    cost  = 1/(2*m) * (np.dot(error.T,error))
    return cost
```

```python
def gradient_descent(X,y,weights,bias,alpha,iterations):
    predictions = np.dot(X,weights) + bias
    error = predictions -y
    cost_history = np.zeros(iterations)
    dw = 1/m * (np.dot(X.T,error))
    db = 1/m * (np.sum(error))
    for i in range(iterations):
        weights -= alpha * dw
        bias -= alpha*db
        cost_history[i] = compute_cost(X,y,weights,bias)
        if i % 100 == 0:  # Print every 100 iterations
            print(f"Iteration {i}: Cost {cost_history[i]}")

    return weights,bias,cost_history
```

```python
weights, bias, cost_history = gradient_descent(X,y,weights,bias,lr,iterations)
print(f"weights :\n {weights}")
print(f"bias : \n {bias}")
```

```python
plt.plot(range(iterations),cost_history,color="blue")
plt.title("Cost function history")
plt.xlabel("number of iterations")
plt.ylabel("cost")
plt.show()
```
